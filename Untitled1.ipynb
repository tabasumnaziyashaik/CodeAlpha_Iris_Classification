```python
# CodeAlpha Task-2: Unemployment Analysis with Python
# Tabasum Naziya Shaik

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Region': ['Andhra Pradesh', 'Bihar', 'Delhi', 'Gujarat', 'Haryana', 'Karnataka', 'Maharashtra', 'Punjab', 'Rajasthan', 'Tamil Nadu', 'Uttar Pradesh', 'West Bengal'] * 3,
    'Date': ['30-01-2020']*12 + ['29-02-2020']*12 + ['31-03-2020']*12,
    'Estimated Unemployment Rate (%)': [5.6, 10.8, 8.9, 6.7, 26.4, 3.4, 5.8, 10.5, 11.9, 4.9, 9.0, 7.6,
                                         5.8, 11.2, 8.7, 6.5, 28.1, 3.6, 5.9, 10.9, 12.1, 5.1, 9.1, 7.9,
                                         5.9, 15.4, 16.7, 8.5, 34.1, 4.2, 8.1, 11.6, 15.2, 6.3, 11.4, 8.2],
    'Estimated Employed': [10000000]*36
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())

plt.figure(figsize=(14, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=df)
plt.title('Unemployment Rate Trends - Covid Lockdown Impact Visible', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.axvline(pd.to_datetime('2020-03-24'), color='red', linestyle='--', label='Lockdown Started')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

march_data = df[df['Date'].dt.month == 3]
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=march_data)
plt.title('Unemployment Rate in March 2020 - Lockdown Started')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print("\n📊 KEY INSIGHTS:")
print("1. Lockdown March 2020 ke baad unemployment spike hua")
print("2. Haryana mein sabse zyada 34.1% unemployment tha")
print("3. Karnataka mein sabse kam 4.2% unemployment tha")
print("4. Clear seasonal trend: March mein sab states mein jump")
```
